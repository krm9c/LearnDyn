```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_0 = np.random.normal(loc=0.0, scale=5.0, size = (10000,10)) 
```

```python
Y_0 = np.exp(1.1*x_0)
```

```python
list_x = []
list_y = []
dataset = {}
a = 0.99
x_prev = x_0
y_prev = np.exp(-a*x_prev)

for i in range(20):
    x_next = np.exp(-a*x_prev) + np.random.normal(loc=0.1, scale=1, size = (10000,10)) 
    y_next = np.sin(a*x_next)
    list_x.append(x_next)
    list_y.append(y_next)
    
    dataset['time'+str(i)] = (x_next, y_next)
    
    x_prev = x_next
```

```python
mean = []
var  = []
time = []
for i in range(len(list_x)):
    time.append(i)
    mean.append(np.mean(list_x[i]))
    var.append(np.std(list_x[i]) )

mean = np.asarray(mean).reshape([-1])
time = np.asarray(time).reshape([-1])
var = np.asarray(var).reshape([-1])
```

```python
print( mean[0] )
```

```python

print(mean.shape, var.shape, time.shape)
plt.clf()
plt.figure(figsize=(40, 10))
plt.plot(time, mean)
plt.fill_between(time, mean-var, mean+var, alpha=0.6, edgecolor='#1B2ACC', facecolor='#FF9848',
    linewidth=4, linestyle='dashdot', antialiased=True)
plt.show()
```

```python
x    =  mean[1000:1100]
t    =  time[1000:1100]
err  =  var [1000:1100]
```

```python
print(mean.shape, var.shape, time.shape)
plt.clf()
plt.figure(figsize=(20, 4))
plt.plot(time, mean)
plt.fill_between(t, x-err, x+err, alpha=1, edgecolor='#1B2ACC', facecolor='#FF9848',
    linewidth=4, linestyle='dashdot', antialiased=True)
plt.show()
```
